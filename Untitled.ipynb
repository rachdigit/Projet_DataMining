```python
categorical_columns = data.select_dtypes[include=[object ,int ]]
```

```python
import matplotlib.pyplot as plt

# Liste des colonnes catégorielles que vous souhaitez analyser

# Créer un graphique pour chaque colonne catégorielle
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Répartition de la colonne: {column}')
    plt.ylabel('')  # Pour enlever l'étiquette de l'axe y
    plt.show()
```

```python
# Sélectionner les colonnes de type 'object' (catégorielles)
categorical_columns = data.select_dtypes(include=['object']).columns

# Créer un graphique en camembert pour chaque colonne catégorielle
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Répartition de la colonne: {column}')
    plt.ylabel('')  # Pour enlever l'étiquette de l'axe y
    plt.show()
```

```python
# Sélectionner les colonnes de type 'object' ou 'int'
categorical_columns = data.select_dtypes(include=['object', 'int']).columns

# Créer un graphique en camembert pour chaque colonne catégorielle
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Répartition de la colonne: {column}')
    plt.ylabel('')  # Pour enlever l'étiquette de l'axe y
    plt.show()
```
