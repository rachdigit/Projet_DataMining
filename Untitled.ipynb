```python
import matplotlib.pyplot as plt

# Liste des colonnes catÃ©gorielles que vous souhaitez analyser

# CrÃ©er un graphique pour chaque colonne catÃ©gorielle
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'RÃ©partition de la colonne: {column}')
    plt.ylabel('')  # Pour enlever l'Ã©tiquette de l'axe y
    plt.show()
```

```python
# SÃ©lectionner les colonnes de type 'object' (catÃ©gorielles)
categorical_columns = data.select_dtypes(include=['object']).columns

# CrÃ©er un graphique en camembert pour chaque colonne catÃ©gorielle
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'RÃ©partition de la colonne: {column}')
    plt.ylabel('')  # Pour enlever l'Ã©tiquette de l'axe y
    plt.show()
```

```python
# SÃ©lectionner les colonnes de type 'object' ou 'int'
categorical_columns = data.select_dtypes(include=['object', 'int']).columns

# CrÃ©er un graphique en camembert pour chaque colonne catÃ©gorielle
for column in categorical_columns:
    plt.figure(figsize=(8, 8))
    data[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'RÃ©partition de la colonne: {column}')
    plt.ylabel('')  # Pour enlever l'Ã©tiquette de l'axe y
    plt.show()
```
